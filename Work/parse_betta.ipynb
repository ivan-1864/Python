```python
import os
import numpy as np
import pandas as pd
lst=[]
name=[]
for root, dirs, files in os.walk("sm"):  
    for filename in files:
        data=pd.read_csv('sm/'+filename, sep='\s+')
        alpha1=data['~AlphaE[arc.min]']
        alpha2=data['~AlphaN[arc.min]']
        print(filename[:6], min(alpha1), max(alpha1),min(alpha2), max(alpha2))
        lst.append([min(alpha1),max(alpha1),min(alpha2),max(alpha2)])
        name.append(int(filename[:6]))
col=['alpha1_min','alpha1_max','alpha2_min','alpha2_max']
df=pd.DataFrame(lst, index=name, columns=col)
df.sort_index(inplace=True)
df.to_excel("Журнал угловых ошибок.xlsx", sheet_name='Smoothed')
```
